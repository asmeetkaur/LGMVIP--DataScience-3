```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Terrorism dataset
terrorism_data = pd.read_csv(r"C:\Users\lenovo\OneDrive\Desktop\lets grow more\globalterrorismdb_0718dist.csv.zip", encoding='ISO-8859-1', low_memory=False)
```

```python
# Display basic information about the dataset
print("Dataset Information:")
print(terrorism_data.info())
```

```python
# Display the first few rows of the dataset
print("\nFirst Few Rows of the Dataset:")
print(terrorism_data.head())
```

```python
# Check for missing values
print("\nMissing Values:")
print(terrorism_data.isnull().sum())
```

```python
# Explore the most affected regions
plt.figure(figsize=(15, 8))
sns.countplot(x='region_txt', data=terrorism_data, order=terrorism_data['region_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Terrorist Attacks by Region')
plt.show()
```

```python
# Explore the most affected countries
plt.figure(figsize=(15, 8))
sns.countplot(x='country_txt', data=terrorism_data, order=terrorism_data['country_txt'].value_counts().index[:15])
plt.xticks(rotation=90)
plt.title('Top 15 Countries with the Most Terrorist Attacks')
plt.show()
```

```python
# Explore the most targeted areas (cities)
plt.figure(figsize=(15, 8))
sns.countplot(x='city', data=terrorism_data, order=terrorism_data['city'].value_counts().index[:15])
plt.xticks(rotation=90)
plt.title('Top 15 Cities with the Most Terrorist Attacks')
plt.show()
```

```python
# Explore attack types
plt.figure(figsize=(12, 6))
sns.countplot(x='attacktype1_txt', data=terrorism_data, order=terrorism_data['attacktype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Distribution of Attack Types')
plt.show()
```
